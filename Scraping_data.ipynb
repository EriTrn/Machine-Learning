```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://id.wikipedia.org/wiki/Daftar_kota_di_Indonesia_menurut_jumlah_penduduk'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
kota = table.find_all('th')
```

```python
kota_tabel = [title.text.strip() for title in kota]

print(kota_tabel)
```

```python
df = pd.DataFrame(columns = kota_tabel)
```

```python
kolom = table.find_all('tr')
```

```python
for row in kolom[1:]:
    data_baris = row.find_all('td')
    baris = [data.text.strip() for data in data_baris]

    length = len(df)
    df.loc[length] = baris
```
