```python
import folium
import numpy as np

# Koordinat pusat Ciamis, Jawa Barat
center_lat, center_lon = -7.5, 108.54

# Membuat objek peta dengan Folium
map_ciamis = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Generate beberapa titik acak di sekitar pusat Ciamis
num_points = 50
latitudes = np.random.uniform(-7.492, -7.502, num_points)
longitudes = np.random.uniform(108.538, 108.548, num_points)

# Tambahkan titik acak ke peta sebagai marker
for lat, lon in zip(latitudes, longitudes):
    folium.Marker(location=[lat, lon], icon=folium.Icon(color="blue")).add_to(map_ciamis)

# Menyimpan hasil peta ke file HTML untuk ditampilkan di browser
map_ciamis.save("map_ciamis.html")
```

```python
import folium
from IPython.display import display

map_center = -7.502, 108.548

map_banjaranyar = folium.Map(location=map_center, zoom_start=12)

folium.Marker(
    [-7.502, 108.548],
    popup="Banjaranyar",
    icon=folium.Icon(color="blue")).add_to(map_banjaranyar)

display(map_banjaranyar)
```
